```python
import pandas as pd

def filter_keep_only_Tg(data):
    """
    Remove todas as colunas das propriedades fornecidas, exceto a coluna 'Tliquidus'.
    """
    properties = ['Tg', 'Tliquidus', 'YoungModulus', 'CTEbelowTg', 'CTE328K',
                  'CTE373K', 'CTE433K', 'CTE483K', 'CTE623K', 'RefractiveIndex', 'AbbeNum']

    # Verifica se 'Tliquidus' está na lista de propriedades e, se estiver, mantém apenas ela
    properties_to_remove = [prop for prop in properties if prop != 'RefractiveIndex']

    # Remove as colunas relacionadas a essas propriedades, exceto 'Tg'
    df_filtered = data.drop(columns=properties_to_remove, errors='ignore')

    return df_filtered

# Exemplo de uso
df = filter_keep_only_Tg(oxide)
print(df)
```

```python
import pandas as pd

def remove_nan_rows(data, column):

    df_filtered = data.dropna(subset=[column])

    return df_filtered

# Exemplo de uso
df = remove_nan_rows(df, 'RefractiveIndex')
print(df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir os bins
bins = np.linspace(1, 2.75, 50)

# Criar o histograma com contagens absolutas
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(df['RefractiveIndex'], bins=bins, alpha=0.6, color='g', edgecolor='black')

# Ajustar curva normal com mediana e MAD
median = np.median(df['RefractiveIndex'])
mad = np.median(np.abs(df['RefractiveIndex'] - median))
robust_std = mad * 1.4826  # fator para converter MAD em std

# Gerar a curva normal ajustada e escalá-la pelo total de composições e largura dos bins
x = np.linspace(plt.xlim()[0], plt.xlim()[1], 100)
bin_width = bins[1] - bins[0]
total = len(df['RefractiveIndex'])
p = norm.pdf(x, loc=median, scale=robust_std) * total * bin_width  # escala para coincidir com a contagem

plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal (Robust)')

# Marcar média e mediana
mu, std = norm.fit(df['RefractiveIndex'])
plt.axvline(mu, color='blue', linestyle='--', label=f'Mean: {mu:.3f}')
plt.axvline(median, color='red', linestyle='--', label=f'Median: {median:.3f}')

plt.xlabel("RefractiveIndex")
plt.ylabel("Number of Compositions")
plt.legend()
plt.show()
```
